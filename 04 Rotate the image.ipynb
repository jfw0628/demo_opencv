# Rotate image 

We can rotate the image 90 degree or 90 degree anticlockwise or 180 degree with a single function.
The complex method will be introduce next.

It is quite simple function, 
cv2.rotate with two params,
- first one is the image(read by opencv)
- second one is the degree we want

If it rotate 90 degree, the resolution of picture will changed

let have a try!

```python
import cv2
import matplotlib.pyplot as plt
figure = plt.figure(figsize=(12, 10))

orig_img = cv2.imread('test2.png')
rotate_90_img = cv2.rotate(orig_img, cv2.ROTATE_90_CLOCKWISE)
anti_rotate_90_img = cv2.rotate(orig_img, cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180_img = cv2.rotate(orig_img, cv2.ROTATE_180)


figure.add_subplot(2, 2, 1) 
plt.imshow(orig_img)
plt.title('Origin image')


figure.add_subplot(2, 2, 2) 
plt.imshow(rotate_90_img)
plt.title('Rotate 90 image')


figure.add_subplot(2, 2, 3) 
plt.imshow(rotate_180_img)
plt.title('Rotate 180 image')



figure.add_subplot(2, 2, 4) 
plt.imshow(rotate_90_img)
plt.title('Anticlockwise Rotate 90 image')


plt.show()
```
