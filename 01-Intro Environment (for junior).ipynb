```python
import requests
res = requests.get("https://cdn.pixabay.com/photo/2014/06/03/19/38/road-sign-361514_1280.png")
with open('test.png', 'wb') as output:
    output.write(res.content)
```

```python
!ls -l test.png 
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# read the image, and display the image size
image = cv2.imread('test.png')
image.shape
```

```python
plt.imshow(image)
```

#### you will find the color is not correct, because it is a bgr format

```python
#change to bgr to rgb and display it
rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```
