# resize image  
## basic method:
- array manipulation, remove and add to resize 
- opencv resize 

```python
# download the test image 
import requests
image_url = "https://www.packedparty.com/cdn/shop/files/0037_1-PP-Sum23-FlowerShopRuler1-4x5_1200x1500.jpg?v=1685626872"
res = requests.get(image_url)
with open('test2.png', 'wb') as output:
    output.write(res.content)
```

```python
import cv2
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 10))
img = cv2.imread('test2.png')
img = img[200:-200, 50:-50, :] #remove some white area
img = img[:,:,::-1]
plt.imshow(img)
```

### Use array manuplation 
Reduce element to zoom out the image


```python
print(img.shape)
plt.figure(figsize=(10, 10))
half_img = img[::2, ::2, :]
print(half_img.shape)
plt.imshow(half_img)
```

### Use opencv resize
https://docs.opencv.org/4.8.0/da/d54/group__imgproc__transform.html#ga47a974309e9102f5f08231edc7e7529d




there are params:  
src, dsize[, dst[, fx[, fy[, interpolation]]]]  
src, the image you want to resize  
dsize, the size of output image, **it is heigh * width, not the same as shape output**  
fx, fy, optional, mean that ratio of output file's size between the src's size  
interpolation, means which method to create new element  

```python
new_size = (550, 550) # height, width 
plt.figure(figsize=(10, 8))
half2_img = cv2.resize(img, new_size)
print(half2_img.shape)
plt.imshow(half2_img)
```

```python
new_size = (2200, 2200) # height, width 
plt.figure(figsize=(15, 10))
double_img = cv2.resize(img, new_size)
print(double_img.shape)
plt.imshow(double_img)
```

let opencv auto calc

```python
new_size = (0, 0) # height, width 
plt.figure(figsize=(15, 10))
double_img = cv2.resize(img, new_size, fx=2.2, fy=2.2)
print(double_img.shape)
plt.imshow(double_img)
```

use different interpolation method

```python
new_size = (0, 0) # height, width 
plt.figure(figsize=(15, 10))
double_img = cv2.resize(img, new_size, fx=1.2, fy=1.2, interpolation=cv2.INTER_CUBIC)
print(double_img.shape)
plt.imshow(double_img)
```
